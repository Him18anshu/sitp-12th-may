```python
pin = "6902" 
balance = 39000 print("ATM") # Ask user for PIN 
entered_pin = input("Enter your PIN: ") 
if entered_pin == pin:    
    while True:        
        print("\n1. check balance")        
        print("2. deposit money")        
        print("3. withdraw money")        
        print("4. close")        
        choose = input("Choose an option: ")# Check balance        
        if choose == "1":            
            print(f"current balance is: {balance}")# Deposit money        
            elif choose == "2":            amount = float(input("Enter amount to deposit: ₹"))            
            balance += amount            
            print(f"₹{amount} deposited successfully.")            
            print(f"New balance: ₹{balance}")# Withdraw money        
            elif choose == "3":            
            amount = float(input("Enter amount to withdraw: "))            
            if amount <= balance:                balance -= amount                
            print(f"₹{amount} withdrawn successfully.")                
            print(f"Remaining balance: {balance}")            
            else:                
            print("Insufficient balance!")# Exit        
            elif choose == "4":            
            print("Thank you")            
            break        
        else:            
print("Invalid option")
else: print("wrong PIN")
```
