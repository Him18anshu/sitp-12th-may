| Mode   | Meaning                         |
| ------ | ------------------------------- |
| `"r"`  | Read file                       |
| `"w"`  | Write file (overwrite old data) |
| `"a"`  | Append data at end              |
| `"x"`  | Create new file                 |

```python
file =open("poit.txt", 'r')
```

```python
file.read()
```

```python
# file.readline()
# file.readlines()
file.close()
```

```python
with open('poit.txt') as file:
    print(file.read())
```

```python
file =open('poit.txt', 'w')
file.write("hiii how are you ? ")
file.close()
file =open('poit.txt', 'r')
file.read()


```

```python
file =open('poit.txt', 'a')
file.write("File Handling in Python means working with files like creating, reading, writing, updating, and deleting files using Python code.")
file.close()
```
